## Обращение к строчкам и столбцам
В DataFrame есть несколько способов для обращения к строкам, столбцам и отдельным элементам таблицы: квадратные скобки `[]` и методы `loc`, `iloc`.

В метод `loc` можно передать значение индекса (число, которое стоит в колонке index) строки, чтобы получить эту строку. С помощью `iloc` тоже можно делать срезы, но в них последний элемент не включается

```python
data3.loc[2] #конкретная строчка
data3.loc[2:4] #срез таблицы
```

```python
data3.iloc[2:4] #срез таблицы, но последний элемент не включается
```

```python
data3.iloc[2:4, 2:6] #выбор и строчек, и столбцов (не включается последняя строчка и последний столбец)
```

```python
#через квадратные скобки можно обращаться к столбцам
data3[['Рейтинг', 'Осенний курс по выбору, приоритет 1']]
#обращение к столбцам по именам
data3.loc[2:4, 'Рейтинг':'Осенний курс по выбору, приоритет 1']
```

## Группировка данных и расчеты
Для разведочного анализа может помочь метод `groupby(column)`.

Он группирует объекты по указанной(-ым) колонке(-ам). Необходимо также указать какую статистику для группировки выводить. Это может быть количество (count), среднее (mean) или другие.

```python
data3.groupby('Осенний курс по выбору, приоритет 2').count()[['ID', 'МИ?']]
#сгруппировали по осеннему курсу, посчитали количество человек (ID) и количество людей с определенной программы (МИ?)
```

```python
data3.groupby('Осенний курс по выбору, приоритет 2').count()[['ID', 'МИ?']].sum(axis=1)
#выводит сумму по количеству всех студентов
```

### Расчетные функции



*   `value_counts()` - посчитать количество наблюдений в зависимости от другой переменной
*   `drop_duplicates()` - удаление дубликатов по определенному столбцу
*   `sort_values()` - сортировка значений
*   `corr()` - расчет корреляции
*   `nunique()` - количество уникальных значений в определенном столбце


```python
#Выводит количество респондентов в зависимости от количества детей. У 274 респондентов нет детей.
df['Children'].value_counts()
```

```python
#поиск дубликатов, необходимо указать столбец
df.drop_duplicates('ID', keep='first', inplace=True)
#проверка на то, что каждый ID-шник уникальный
print(all(df['ID'].value_counts() == 1))
```

```python
#сортировка по определенному столбцу
df.sort_values('ID', ascending=False, inplace=True)
```

```python
#корреляционная матрица
corrmat = df[['Income', 'Children', 'Cars', 'Age']].corr()
print(corrmat)
plt.figure(figsize=(8, 8))
sns.heatmap(corrmat, square=True)
plt.show()
```

## Соединение таблиц данных

Функция `pd.concat` используется для объединения датафреймов `data1_1` и `data1_2` по оси 1, то есть путем добавления столбцов из `data1_2` к `data1_1`.

Функции `pd.merge` используются для объединения датафреймов `data1_1` и `data1_2` по столбцу 'X1' с различными типами объединения (left, right, inner, outer).

Параметр 'how' определяет тип объединения:
- **left** - возвращает объединение по значениям 'X1' из первого датафрейма `data1_1`
- **right** - возвращает объединение по значениям 'X1' из второго датафрейма `data1_2`
- **inner** - возвращает объединение только по общим значениям 'X1'
- **outer** - возвращает объединение по всем значениям 'X1'

```python
data1_1 = pd.DataFrame({
    "X1" : ["A", "B", "C", "D"],
    "X2" : [1, 2, 3, 4]
})

data1_2 = pd.DataFrame({
    "X1" : ["A", "B", "C", "E"],
    "X3" : [1, 0, 1, 0]
})
```

```python
pd.concat([data1_1,data1_2], axis=1) #соединение невзирая на значения
```

```python
pd.merge(data1_1, data1_2, how='left', on='X1')
```

```python
pd.merge(data1_1, data1_2, how='right', on='X1')
```

```python
pd.merge(data1_1, data1_2, how='inner', on='X1')
```

```python
pd.merge(data1_1, data1_2, how='outer', on='X1')
```
